## segregating the director name from req_dataset


```python
def get_director(x):
    if '(director)' in x:
        return x.split('(director)')[0]
    elif '(directors)' in x:
        return x.split('(directors)')[0]
    else:
        return x.split('(director/screenplay)')[0]
```

```python
req_dataset['director_name']=req_dataset['Cast and crew'].apply(get_director)
```

```python
def get_actor_1(x):
    if('screenplay' not in x):
        if 'director' in x:
            return x.split('(director);')[-1].split(',')[0]
        if 'directors' in x:
            return x.split('(directors);')[-1].split(',')[0]
        
    return (x.split('screenplay);')[-1].split(',')[0])
```

```python
req_dataset['actor_1_name']=req_dataset['Cast and crew'].apply(get_actor_1)
```

```python
def get_actor_2(x):
    if len(x.split('screenplay);')[-1].split(','))<2:
        return np.NaN
    else:
        return x.split('screenplay);')[-1].split(',')[1]
        
```

```python
req_dataset['actor_2_name']=req_dataset['Cast and crew'].apply(get_actor_2)
```

```python
def get_actor_3(x):
    if len(x.split('screenplay);')[-1].split(','))<3:
        return np.NaN
    else:
        return x.split('screenplay);')[-1].split(',')[2]
```

```python
req_dataset['actor_3_name']=req_dataset['Cast and crew'].apply(get_actor_3)
```

```python
req_dataset.drop(['Cast and crew'],axis=1)
```

```python
 def remove_codirector(x):
    if 'co-director' in x:
            return x.split('(co-director);')[-1].split(',')[0]
    return x
```

```python
req_dataset['actor_1_name']=req_dataset['actor_1_name'].apply(remove_codirector)
```

```python
req_dataset
```

```python
req_dataset.drop(['Cast and crew'],axis=1,inplace=True)
```

```python
from tmdbv3api import Movie
tmdb_movie=Movie()
```

```python
import json
```

```python
def get_genres(x):
    genres=[]
    res=tmdb_movie.search(x)
    movie_id=res[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    response_json_value=response.json()
    if response_json_value['genres']:
        for i in range(0,len(response_json_value['genres'])):
            genres.append(response_json_value['genres'][i]['name'])
        return " ".join(genres)
    else:
        return np.NaN
    
    
```

```python
get_genres('avatar')
```

```python
req_dataset['genres']=req_dataset['Title'].apply(get_genres)
```
